```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
ab_cost = pd.read_csv('./datasets/abortion_costs_simplified_eks.csv')
```

```python
ab_cost.head(10)
```

```python
ab_cost.loc[51, 'county'] = 'Trinity'

ab_cost.loc[51, 'total_expenditures'] = '0.0'
```

```python
ab_count = pd.read_csv('./datasets/abortion_counts_simplified_eks.csv')
```

```python
ab_count.head(10)
```

# Merge datasets

```python
ab_combined = pd.merge(ab_count, ab_cost, on = 'county', how = "inner")
```

```python
ab_combined.shape
```

```python
ab_combined.isnull().sum()
```

```python
ab_combined
```

```python
ab_combined.loc[44, 'abortion_rs_count_total'] = '0.0'
```

```python
ab_combined['total_expenditures'] = ab_combined['total_expenditures'].str.replace('$', '')
```

```python
ab_combined['total_expenditures'] = ab_combined['total_expenditures'].str.replace(',', '')
```

```python
ab_combined
```

# Create X and y

```python
X = ab_combined.drop(columns = ['county', 'total_expenditures'])
y = ab_combined['total_expenditures']
```

```python
# instantiate the model
lr = LinearRegression()
# fit the model
lr.fit(X, y)
# score the model - this just is what it is, there's not a lot we can do to fix it, but it's something that will print out so you know everything worked
lr.score(X, y)
```
