# Scale and create poly features and check the corr

```python
#read in data
df = pd.read_csv('./ca_dropout_and_predictors_v5.csv')
```

```python
#drop the unnamed column and get set index to county
df = df.drop(columns = 'Unnamed: 0').set_index('county')
```

```python
X = df.drop(columns = ['dropout_rate_2019_cohort', 'abortion_rs_count_total', 'daycare_slots_infant_facility_capacity', 'daycare_slots_infant_rate_slots', 'daycare_slots_infant_num_facility', 'asthma_number_of_deaths', 'asthma_age_adjusted_mortality_rate', 'dropout_sqrt', 'dropout_quar_rt', 'dropout_3rd_rt', 'dropout_not_zero', 'dropout_log'])
y = df[['dropout_log']]
```

```python
poly = PolynomialFeatures(include_bias=False)
```

```python
X_poly = poly.fit_transform(X)
```

```python
poly_df = pd.DataFrame(X_poly, columns = poly.get_feature_names_out())
```

```python
# get column name back out and concat on index with poly
df_county = pd.read_csv('./ca_dropout_and_predictors_v5.csv')[['county']]
corr_df = pd.concat([poly_df, df_county], axis = 1)
```

```python
#redifne X and y and put the county back as the index for X
X = corr_df.set_index('county')
y = df[['dropout_log']]
```

```python
#make a corr df that i can run correlation on with the dropout log column
corr_df = pd.merge(corr_df, y, left_on = 'county', right_on = 'county')
corr_df = corr_df.set_index('county')
```

```python
pd.set_option('display.max_columns', 100)
corr = corr_df.corr(numeric_only = True)[['dropout_log']].sort_values(by = 'dropout_log', ascending = False)
```

```python
corr['abs_value_dropout_log'] = np.abs(corr['dropout_log'])
```

```python
# gives me the 10 best correlated features to add to the model
corr.sort_values(by = 'abs_value_dropout_log', ascending = False).head(11)
```

```python
corr.to_csv('./corr_poly.csv')
corr_df.to_csv('./df_poly.csv')
```
