```python
# import the packages
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# read the data in
data = pd.read_csv('./datasets/ca_dropout_and_predictors.csv')
```

```python
data.head()
```

```python
# subset it - the exact variable names will be different for each bullet point
data2 = data[['abortion_rs_count_total', 'abortion_medicaid_expenditures', 'num_of_indiv_A20']]
```

```python
data2.head()
```

```python
# drop any remaining NAs
data2.dropna()
# create the target variable - this variable name will be different each time too
```

```python
y = data2['abortion_rs_count_total']
# create the predictor matrix - these variable names will be different each time too
X = data2[['abortion_medicaid_expenditures', 'num_of_indiv_A20']]
```

```python
# instantiate the model
lr = LinearRegression()
# fit the model
lr.fit(X, y)
# score the model - this just is what it is, there's not a lot we can do to fix it, but it's something that will print out so you know everything worked
lr.score(X, y)
```
